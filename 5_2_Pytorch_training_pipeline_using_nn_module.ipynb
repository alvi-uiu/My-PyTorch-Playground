```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
```

```python
df.head()
```

### train test split

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

### scaling

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
y_train
```

### Label Encoding

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

### Numpy arrays to PyTorch tensors

```python
X_train_tensor = torch.from_numpy(X_train.astype(np.float32))
X_test_tensor = torch.from_numpy(X_test.astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.astype(np.float32))
```

```python
X_train_tensor.shape
```

```python
y_train_tensor.shape
```

### Defining the model

```python
import torch.nn as nn


class MySimpleNN(nn.Module):

  def __init__(self, num_features):  # Constructor to initialize the model
    super().__init__()  # Calls the parent class (nn.Module) constructor, necessary for PyTorch models

    # Define the neural network layers using nn.Sequential
    self.network = nn.Sequential(
        nn.Linear(num_features, 64),  # First hidden layer: Input size = num_features, Output size = 64
        nn.ReLU(),  # ReLU activation function for non-linearity
        nn.Linear(64, 32),  # Second hidden layer: Input size = 64, Output size = 32
        nn.ReLU(),  # ReLU activation for non-linearity
        nn.Linear(32, 1),  # Output layer: Input size = 32, Output size = 1
        nn.Sigmoid()  # Sigmoid activation for binary classification output
    )

  def forward(self, features):  # Forward method defines how the input flows through the network
    out = self.network(features)  # Pass input through the network layers defined in nn.Sequential
    return out  # Return the final output
```

### Important Parameters

```python
learning_rate = 0.1
epochs = 25
```

```python
# define loss function
loss_function = nn.BCELoss() # binary cross entropy loss
```

### Training Pipeline

```python
# Create model
model = MySimpleNN(X_train_tensor.shape[1])  # Initialize the model with input size based on training data

# Define optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)  # Using Stochastic Gradient Descent (SGD) with specified learning rate

# Define loop for training the model
for epoch in range(epochs):  # Iterate through the number of epochs

  # Forward pass
  y_pred = model(X_train_tensor)  # Pass input features through the model to get predictions

  # Calculate loss
  loss = loss_function(y_pred, y_train_tensor.view(-1, 1))  # Compute the loss using the model's predictions and true labels

  # Clear gradients
  optimizer.zero_grad()  # Reset gradients to zero to prevent accumulation from previous iterations

  # Backward pass
  loss.backward()  # Compute the gradients of the loss with respect to the model's parameters

  # Parameters update
  optimizer.step()  # Update the model's parameters based on the computed gradients

  # Print loss for each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')  # Display the current epoch and the loss value
```

### Evaluation

```python
# model evaluation
with torch.no_grad():
  y_pred = model.forward(X_test_tensor)
  y_pred = (y_pred > 0.5).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
